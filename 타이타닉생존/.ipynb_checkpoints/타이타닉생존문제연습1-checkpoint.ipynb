### 항목 정리
1. passenger ID : 승객 번호
2. Survived : 생존은 1, 사망은 0 ( train파일에는 답이 주어져 있고, test파일은 빈값)
3. Pclass : 선실 등급 (1 : 1st , 2 : 2nd, 3 : 3rd)
4. Sex : 성별 male/female
5. Age : 나이, 이 항목은 NaN이 많음, 이를 처리하는 것이 중요
6. Sibsp : 형제, 자매의 수/ 배우자 등이 승선한 경우 수
7. Parch : 부모나 자식과 같이 탄 경우 수(부모+아이의 가족 탑승자 수)
8. Ticket : 표 번호
9. Fare : 운임
10. Cabin : 선실 번호, 여기도 NaN이 많음
11. embarked : 승선한 항구 (C : Cherbourg, Q : Queenstown, S : Southampton)
